```python
X = df.drop(columns = ['price'],axis = 1)
y = df['price']
```

```python
X
```

# One Hot Encoding

```python
df.columns
```

```python
X_one_hot_encode = pd.get_dummies(X,columns = ['model','transmission','fuelType'],drop_first = True)
```

```python
X_one_hot_encode = X_one_hot_encode.astype(int)
```

```python
X_one_hot_encode
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model', 'transmission', 'fuelType']

X_lable_encode = X.copy()  # make a safe copy
label_encoders = {}

for col in columns:
    le = LabelEncoder()
    X_lable_encode[col] = le.fit_transform(X_lable_encode[col].astype(str))  # Convert to string in case of nulls
    label_encoders[col] = le
```

```python
X_lable_encode
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
numerical_cols = ['year', 'mileage', 'tax', 'mpg', 'engineSize']
scaler = StandardScaler()
X_one_hot_encode[numerical_cols] = scaler.fit_transform(X_one_hot_encode[numerical_cols])
```

```python
X_one_hot_encode
```

```python
X_lable_encode.columns
```

```python
X_lable_encode[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(X_lable_encode[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
X_lable_encode
```

# Model Creation with One Hot Encoded Data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot_encode, y, test_size=0.33, random_state=42)
```

```python
model_ohe = LinearRegression()
model_ohe.fit(X_train,y_train)
```

```python
y_pred_ohe = model_ohe.predict(X_test)
```

```python
y_pred_ohe
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_pred_ohe)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1 - r2) * (n - 1)) / (n - p - 1)
print("Adjusted R Score:", adjusted_r2)
```
